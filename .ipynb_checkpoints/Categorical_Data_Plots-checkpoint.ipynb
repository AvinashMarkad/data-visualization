### Categorical Data Plots
1. BoxPlot
2. Violinplot
3. stripplot
4. swarmplot
5. barplot
6. countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### 6. countplot

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python

sns.countplot(x=df['sex'])
```

 ### 2. Violinplot

```python
sns.violinplot(x='day',y='tip', data=df)
```

### 3. stripplot

```python
sns.stripplot(x='sex', y='tip',data=df)
```

### 4. swarmplot

```python
sns.swarmplot(x='sex', y='tip', data=df) 
```

### 5. barplot

```python
import numpy as np
sns.barplot(x=df['sex'], y=df['total_bill'], estimator = np.sum)
```

### 1. BoxPlot

```python
sns.boxplot(
    x=df['tip'], 
    y=df['day'], 
    hue=df['day'],
    palette="rainbow", 
    legend=True
)
```
