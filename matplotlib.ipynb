```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlin basics

```python
x = np.linspace(-11,5,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

subplots in matplots

```python
plt.subplot(2,2,1)
plt.plot(x+x,y+x)
plt.subplot(2,2,2)
plt.plot(x-y,y-x)
plt.subplot(2,2,3)
plt.plot(x*x,y*y)
plt.subplot(2,2,4)
plt.plot(x/x,y/x)
```

object oriented way

```python
fig = plt.figure(figsize=(10,5),dpi=300)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.5,0.5,.5,.5])
axis2.plot(x*x,y*y)
```

types of plots

```python
plt.scatter(x,y)
```

```python
plt.hist([x,y])
```

```python
plt.boxplot(x)
```

saving plots

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
plt.savefig("Save Plot.png")
```

working with images

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('image-cat.jpg')
plt.imshow(img)
#plt.axis('off')
```

```python
croped_img = img[1000:3000,1000:3000]
plt.imshow(croped_img)
```

designing

```python
plt.plot(x, y, 
         color='red', 
         linestyle='-', 
         linewidth=3, 
         marker='*', 
         markersize=15,
         markerfacecolor='white', 
         markeredgecolor='green')

plt.title("Styled Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
#plt.legend()
plt.grid(True)
plt.show()
```
